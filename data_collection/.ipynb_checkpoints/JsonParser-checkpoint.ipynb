```python
with open("municipalities-topo.json") as json_data:
    data = json.load(json_data)
    for d in data["objects"]["main"]["geometries"]:
        if d["properties"]["ID_1"] != 15:
            del d
    with open('Mapping_Seoul/seoul-topo.json', 'w') as outfile:
        json.dump(data, outfile)
```
